```python
from sklearn import svm, metrics
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import numpy as np
from skimage import img_as_ubyte, io, color
from sklearn.model_selection import train_test_split
from sklearn import svm, metrics
from collections import Counter
```

creating a labels list and an images list so we cant get out trainig data

```python
def import_images(path):
  images = []

  file_name = [file for file in sorted(os.listdir(os.path.join(GOOGLE_DRIVE_PATH, path))) if file.endswith('.jpg')]
  for file in file_name:
    images.append(io.imread(os.path.join(GOOGLE_DRIVE_PATH, path, file)))

  return images;
  
```

```python
X = import_images('CW_Dataset/CW_Dataset/train')
```

```python
np.shape(X)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(10, 5), sharex=True, sharey=True)
ax = axes.ravel()

for i in range(10):
  ax[i].imshow(X[i], cmap='gray')
plt.show()
```

```python
def import_labels(path):
  label= []
  with open ((os.path.join(GOOGLE_DRIVE_PATH, path)), "r") as myfile:
    data = myfile.read().splitlines()
    for i in data:
      x,y = i.split(" ")
      label.append(y)
  return label

```

```python
y = import_labels('CW_Dataset/CW_Dataset/labels/list_label_train.txt')
```

```python
np.shape(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=True, stratify=y)
```

```python

fig, axes = plt.subplots(nrows=2, ncols=5, figsize = (10,5), sharex=True, sharey = True)
ax = axes.ravel()

for i in range(10):
  ax[i].imshow(X_train[i], cmap='gray')
  ax[i].set_title(f'Label: {y_train[i]}')
  
fig.tight_layout()
plt.show()
```
